```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(r"sales_data.csv")
df
```

Line Plot â€” Revenue Over Time

```python
plt.figure(figsize=(20,6))
plt.plot(df['Date'], df['Revenue'], marker='o', color='teal')
plt.title('Monthly Revenue Trend')
plt.xlabel('Date')
plt.ylabel('Revenue ($)')
plt.grid(True)
plt.show()
```

Bar Chart â€” Units Sold by Product

```python
plt.figure(figsize=(8,4))
plt.bar(df['Product'], df['Units Sold'], color='orange')
plt.title('Units Sold by Product')
plt.xlabel('Product')
plt.ylabel('Units Sold')
plt.show()
```

Horizontal Bar Chart â€” Profit by Region

```python
region_profit = df.groupby('Region')['Profit'].sum()
pd.DataFrame(region_profit)
```

```python
region_profit = df.groupby('Region')['Profit'].sum()

plt.figure(figsize=(6,4))
region_profit.plot(kind='barh', color='purple')
plt.title('Total Profit by Region')
plt.xlabel('Profit ($)')
plt.ylabel('Region')
plt.show()
```

Scatter Plot â€” Revenue vs Profit

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Revenue'], df['Profit'], color='darkgreen', s=100, alpha=0.7)
plt.title('Revenue vs Profit')
plt.xlabel('Revenue ($)')
plt.ylabel('Profit ($)')
plt.grid(True)
plt.show()
```

Pie Chart â€” Revenue Contribution by Product

```python
product_revenue = df.groupby('Product')['Revenue'].sum()

plt.figure(figsize=(6,6))
plt.pie(product_revenue, labels=product_revenue.index, autopct='%1.1f%%', startangle=140)
plt.title('Revenue Share by Product')
plt.show()
```

Box Plot â€” Profit Distribution by Region

```python
plt.figure(figsize=(6,4))
df.boxplot(column='Profit', by='Region', grid=False, patch_artist=True)
plt.xlabel('Region')
plt.ylabel('Profit ($)')
plt.show()
```
